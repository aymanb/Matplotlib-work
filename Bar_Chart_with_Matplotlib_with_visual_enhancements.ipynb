Import needed libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.patheffects as path_effects
```



---



create a dataframe for the csv file of ***Karim Benzema*** Goals and games by season with **Real Madrid**

```python
df = pd.read_csv('Benzema_RM.csv')
df
```



---



```python
fig = plt.figure(figsize=(6,2.5), dpi = 200)
ax = plt.subplot(111)

ax.bar(df['Season'], df['Goals'])

ax.tick_params(axis = "x", rotation = 90)
```



---



```python
fig = plt.figure(figsize=(6, 2.5), dpi = 300, facecolor = "#979dac")
ax = plt.subplot(111, facecolor = "#979dac")

#spines: taking off the spines from top and right
ax.spines["top"].set(visible = False)
ax.spines["right"].set(visible = False)

#add grid, background color, and Y axis label
ax.grid(True, color = "lightgrey", linestyle = "--")
ax.set_ylabel("Goals in all competitions")

#this section where we create bars with 
ax.bar(
    df['Season'],
    df['Goals'],
    ec = "white",
    lw = .75,
    color = "#001233",
    zorder = 3,
    width = 0.75
)

ax.tick_params(axis = "x", rotation = 90) #this to adjust the rotation of the x axis for Season to fit in

#this section, we add the value label (goals) of each bar/season
for index, y in enumerate(df['Goals']):
    ax.annotate(
        xy = (index, y),
        text = f"{y:}",
        xytext = (0, 7),
        textcoords = "offset points",
        size = 7,
        color = "#001233",
        ha = "center",
        va = "center",
        weight = "bold"
    )
```
